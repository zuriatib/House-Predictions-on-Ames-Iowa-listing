## 1) Encoding

a) I will start with interger encoding for ordinal variablabes. I am defining a numerical encoding, seeing that there are several features (ie, Overall Qual/Cond) that are already encoded as such with values between 1 to 10. 

- Land Slope
- Lot Shape
- Utilities
- Land Slope
- Exter Cond (values with Ex, Gd, TA(average/typical), Fa, Po)
- Bsmt Qual (values with Ex, Gd, TA (typical), Fa, Po, NA)
- Bsmt Exposure (values Gd, Av, Mn, No NA)
- BsmtFin Type1/2 (GLQ,ALQ,BLQ,Rec, LwQ,Unf, NA)
- HeatingQC (Ex, Gd, TA, Fa, Po)
- Electrical (SBrker, FuseA(average),FuseF(fair),FuseP(poor), Mix)
- KitechenQual (Ex, Gd, TA, Fa, Po)
- Functional (Typ, Min1, Min2, Mod, Maj1, Maj2, Sev, Sal)
- FireplaceQu (Ex, Gd, TA, Fa, Po, NA)
- Garage Finish (Fin, RFn, Unf, NA)
- Garage Qual/Garage Cond (Ex, Gd, TA, Fa, Po, NA)
- Paved Drive (Y, P, N)
- Pool QC (Ex, Gd, TA, Fa, NA)
- Fence (GdPrv, MnPrv, GdWo,MnWw,NA)


```python
def encode_cols(df, myList):
    for col in myList:
        if col == 'Lot Shape':
            df[col] = df[col].map({'Reg':4, 'IR1':3, 'IR2':2, 'IR3':1})
        if col == 'Utilities':
            df[col] = df[col].map({'AllPub':4, 'NoSewr':3, 'NoSeWa':2, 'ELO':1})
        if col == 'Land Slope':
            df[col] = df[col].map({'Gtl':3, 'Mod':2,'Sev':1})
        if col == 'Bsmt Exposure':
            df[col] = df[col].map({'Gd':5, 'Av':4, 'Mn':3, 'No':2, 'NA':1})
        if col == 'Electrical':
            df[col] = df[col].map({'SBrkr':5, 'FuseA':4,'FuseF':3,'FuseP':2, 'Mix':1})
        if col == 'Functional':
            df[col] = df[col].map({'Typ':8, 'Min1':7, 'Min2':6, 'Mod':5, 'Maj1':4, 'Maj2':3, 'Sev':2, 'Sal':1})
        if col == 'Garage Finish':
            df[col] = df[col].map({'Fin':4, 'RFn':3, 'Unf':2, 'NA':1})
        if col == 'Paved Drive':
            df[col] = df[col].map({'Y':4,'P':3,'N':2})
        if col == 'Fence':
            df[col] = df[col].map({'GdPrv':5, 'MnPrv':4, 'GdWo':3,'MnWw':2,'NA':1})
        if (col == 'BsmtFin Type 1') | (col == 'BsmtFin Type 2'):
            df[col] = df[col].map({'GLQ':7,'ALQ':6,'BLQ':5,'Rec':4, 'LwQ':3,'Unf':2, 'NA':1})
    return df  
        
def encode_wo_NA(df,myList):
    for col in myList:
        df[col] = df[col].map({'Ex':5, 'Gd':4, 'TA':3, 'Fa':2, 'Po':1})
    return df
        
def encode_w_NA(df,myList):    
    for col in myList:
        df[col] = df[col].map({'Ex':6, 'Gd':5, 'TA':4, 'Fa':3, 'Po':2, 'NA':1})
    return df
```

```python
df['Central Air'] = df['Central Air'].map({'Y':1, 'N':0})
```

```python
df['Pool QC'] = df['Pool QC'].map({'Ex':5, 'Gd':4, 'TA':3, 'Fa':2, 'NA':1})
```

```python
lst1 = ['Lot Shape','Utilities','Land Slope','Bsmt Exposure','Electrical','Functional','Garage Finish','Paved Drive','Fence','BsmtFin Type 1','BsmtFin Type 2']
lst2 = ['Exter Qual','Exter Cond','Heating QC','Kitchen Qual']
lst3 =['Bsmt Qual','Bsmt Cond','Fireplace Qu','Garage Qual','Garage Cond']

df = df.pipe(encode_cols,lst1)
df = df.pipe(encode_wo_NA, lst2)
df = df.pipe(encode_w_NA, lst3)
```

### a) Regularization with Grid Search

```python
#regularization with Ridge Regression
from sklearn.linear_model import Ridge

#define alphas
alphas = np.logspace(-20,10,num=100)

train_scores = []
test_scores = []

for alpha in alphas:
    ridge = Ridge(alpha=alpha)
    ridge.fit(X2_tr,y2_tr)
    
    y2_tr_pred_ridge = ridge.predict(X2_tr)
    train_scores.append(np.sqrt(msle(y2_tr,y2_tr_pred_ridge)))
    
    y2_te_pred_ridge = ridge.predict(X2_te)
    test_scores.append(np.sqrt(msle(y2_te,y2_te_pred_ridge)))
```

```python
# Plot the validation curves
plt.semilogx(alphas, train_scores, label='train curve')
plt.semilogx(alphas, test_scores, label='test curve')
plt.xlabel('alphas')
plt.ylabel('RMSLE')
plt.legend()
plt.show()
```

```python
diff = []
for (train_score,test_score) in zip(train_scores,test_scores):
    val = (np.abs(train_score - test_score))
    diff.append(val)    

idx = (np.argmin(diff))
best_alpha = alphas[idx]
test = test_scores[idx]
train = train_scores[idx]

print('Best alpha value {:.1e} with test score {:.4f} and train score {:.4f}'.format(best_alpha,test,train))
```

### a) Regularization with Grid Search

```python
#define alphas
alphas = np.logspace(-10,10,num=100)

train_scores = []
test_scores = []

for alpha in alphas:
    ridge_com = Ridge(alpha=alpha)
    ridge_com.fit(X3_tr,y3_tr)
    
    
    y3_tr_pred_ridge = ridge_com.predict(X3_tr)
    train_scores.append(np.sqrt(msle(y3_tr,y3_tr_pred_ridge)))
    
    y3_te_pred_ridge = ridge_com.predict(X3_te)
    test_scores.append(np.sqrt(msle(y3_te,y3_te_pred_ridge)))
```

```python
# Plot the validation curves
plt.semilogx(alphas, train_scores, label='train curve')
plt.semilogx(alphas, test_scores, label='test curve')
plt.xlabel('alphas')
plt.ylabel('RMSLE')
plt.legend()
plt.show()
```

```python
diff = []
for (train_score,test_score) in zip(train_scores,test_scores):
    val = (np.abs(train_score - test_score))
    diff.append(val)    

idx = (np.argmin(diff))
best_alpha = alphas[idx]
test = test_scores[idx]
train = train_scores[idx]

print('Best alpha value {:.1e} with test score {:.4f} and train score {:.4f}'.format(best_alpha,test,train))
```
